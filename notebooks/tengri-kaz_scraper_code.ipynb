```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import time

articles = []
base = "https://kaz.tengrinews.kz"

# 1500 pages covers approximately back to early 2024
for page in tqdm(range(1, 1501)):
    url = f"{base}/news/page/{page}/"

    try:
        r = requests.get(url, timeout=15, headers={'User-Agent': 'Mozilla/5.0'})
        if r.status_code != 200:
            break

        soup = BeautifulSoup(r.content, 'html.parser')
        items = soup.select('.content_main_item')

        if not items:
            break

        for item in items:
            try:
                title_tag = item.select_one('.content_main_item_title a')
                if not title_tag: continue

                link = base + title_tag['href']
                title = title_tag.text.strip()

                # Check date to see if we reached 2023
                meta = item.select_one('.content_main_item_meta span')
                date_text = meta.text.strip() if meta else ""

                # STOP if we hit 2023
                if "2023" in date_text:
                    print(f"\nReached end of 2024 data at page {page}. Finalizing...")
                    break # Breaks the 'for item in items' loop

                # Enter article
                ar = requests.get(link, timeout=10, headers={'User-Agent': 'Mozilla/5.0'})
                s = BeautifulSoup(ar.content, 'html.parser')

                tags = [t.text.strip() for t in s.select('.content_main_tag a')]
                text_div = s.select_one('.content_main_text')
                full_text = " ".join([p.text.strip() for p in text_div.find_all('p')]) if text_div else ""

                articles.append({
                    'title': title,
                    'url': link,
                    'tags': ", ".join(tags),
                    'text': full_text,
                    'date': date_text
                })
                time.sleep(0.1)
            except:
                continue

        # This checks if the inner loop broke because of the 2023 date
        if len(articles) > 0 and "2023" in articles[-1]['date']:
            break # Breaks the 'for page in range' loop

        if page % 50 == 0:
            pd.DataFrame(articles).to_csv(f'kaz_checkpoint_{page}.csv', index=False, encoding='utf-8-sig')

    except Exception as e:
        print(f"Error: {e}")
        break

df = pd.DataFrame(articles).drop_duplicates('url')
df.to_csv('tengrinews_kazakh_2024_2026.csv', index=False, encoding='utf-8-sig')
print(f"Finished! Total: {len(df)} articles.")
```
